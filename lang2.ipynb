```python
from langchain_core.prompts.string import PromptTemplateFormat
from langchain_core.prompts import PromptTemplate # PromptTemplate ì„í¬íŠ¸ ì¶”ê°€
# PromptTemplate ì‚¬ìš© : ê¸°ë³¸ í”„ë¡¬í”„íŠ¸ + ë³€ìˆ˜ ì‚½ì…
mytemplate = """ë„ˆëŠ” í•œêµ­ì–´ ì „ë¬¸ê°€ì•¼.
ì•„ë˜ ë‚´ìš©ìœ¼ë¡œ 5í–‰ ì´ë‚´ì˜ ì•„ë¦„ë‹¤ìš´ ì‹œë¥¼ ì‘ì„±í•´ì¤˜.
"{content}"
"""
print("template : ", mytemplate)

prompt = PromptTemplate(
    input_variables=["content"],
    template=mytemplate
)
fill_prompt = prompt.format(content="ê°€ì„í•˜ëŠ˜")
print("fill_prompt :" , fill_prompt)

poem_llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.9)
poem_response = poem_llm.invoke(fill_prompt)
print(poem_response.content) # .content ì†ì„±ì„ ì¶œë ¥í•˜ì—¬ ì¤„ë°”ê¿ˆ ì ìš©
```

```python
# TOOL ì‚¬ìš© (TOOL + AGENT êµ¬ì¡°)
import pprint # pprint ëª¨ë“ˆ ì„í¬íŠ¸
print('ê³„ì‚°ê¸° íˆ´ ì‘ì„±-----------')
from langchain.tools import tool
from langchain.agents import create_agent

# Tool ì •ì˜
@tool
def myCulc(experession:str) -> str:
  """ê°„ë‹¨í•œ ì‚¬ì¹™ì—°ì‚° ìˆ˜ì‹ì„ ê³„ì‚°í•˜ê³  'ìˆ˜ì‹ = ê°’ ' í˜•íƒœì˜ ë¬¸ìì—´ë¡œ ë°˜í™˜í•œë‹¤. """
  try:
    result = eval(experession) # eval("6+7")
    return f"{experession} = {result}"
  except Exception as e:
    return f"ê³„ì‚° ì‹¤íŒ¨ : {e}"

tools = [myCulc]
model = ChatOpenAI(model="gpt-4o-mini", temperature=0.0)

agent = create_agent(
    tools = tools,
    model = model,
    system_prompt = (
        "ë„ˆëŠ” ìˆ˜í•™ ê³„ì‚°ì„ ë„ì™€ì£¼ëŠ” ì–´ì‹œìŠ¤í„´ìŠ¤ì•¼"
        "ê°€ëŠ¥í•˜ë©´ myCulc íˆ´ì„ ì‚¬ìš©í•´ ì •í™•í•œ ê°’ì„ ê³„ì‚°í•´ì¤˜"
        "ì¶œë ¥ì€ í˜•ì‹ì„ ì§€ì¼œì„œ ë‹µí•´. ìˆ˜ì‹ = ê°’"
    ),
)

question = " 7 * (3+2) / 2ëŠ” ì–¼ë§ˆì•¼?"
result = agent.invoke({
    "messages":[
        {"role":"user", "content":question}
    ]
})
# pprint.pprint(result) # pprintë¥¼ ì‚¬ìš©í•˜ì—¬ ë³´ê¸° ì¢‹ê²Œ ì¶œë ¥

last_msg = result["messages"][-1]
print("ìµœì¢… ë‹µë³€ : " , last_msg.content)
```
